#### Based on the sample size, the median salary does not seem to be larger for different majors.<br> There is no trend that shows us that students who majored in subjects that were majority female make more money. <br>There seems to be a trend in number of full time employees and median salary.

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize = (18,10))
for r in range(0,4):
    ax = fig.add_subplot(2,2, r+1)
    ax = recent_grads[cols[r]].hist()
    ax.set_xlabel(cols[r])
    
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize = (18,10))
for r in range(4,8):
    ax = fig.add_subplot(2,2, r-3)
    ax = recent_grads[cols[r]].hist()
    ax.set_xlabel(cols[r])
    
```

#### About 20% of majors are predominantly women. <br> About 20% of majors are predominantly men.<br>Most common median salary range is between 30k and 40k
